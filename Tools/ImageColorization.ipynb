```python

# importing the required module
import matplotlib.pyplot as plt
from decimal import Decimal
import csv


def graphLossResults(file):
    epoch = []
    loss = []
    with open(file,'rt')as f:
        data = csv.reader(f)
        counter = 0
        for row in data:
            if len(row) > 0:
                epoch.append(Decimal(row[0]))
                loss.append(Decimal(row[1]))
        # plotting the points
        plt.plot(epoch, loss)
        
        # naming the x axis
        plt.xlabel('x - EPOCHS')
        # naming the y axis
        plt.ylabel('y - LOSS')
        
        # giving a title to my graph
        plt.title('EPOCHS VS LOSS')
        
        # function to show the plot
        plt.show()

def graphMetricsResults(file):
    epoch = []
    PSNR = []
    SSIM = []
    with open(file,'rt')as f:
        data = csv.reader(f)
        counter = 0
        for row in data:
            if len(row) > 0:
                epoch.append(Decimal(row[0]))
                PSNR.append(Decimal(row[1]))
                SSIM.append(Decimal(row[2]))
        # plotting the points
        plt.plot(epoch, PSNR)
        
        # naming the x axis
        plt.xlabel('x - EPOCHS')
        # naming the y axis
        plt.ylabel('y - PSNR')
        
        # giving a title to my graph
        plt.title('EPOCHS VS PSNR')
        
        # function to show the plot
        plt.show()
        
        # plotting the points
        plt.plot(epoch, SSIM)
        
        # naming the x axis
        plt.xlabel('x - EPOCHS')
        # naming the y axis
        plt.ylabel('y - SSIM')
        
        # giving a title to my graph
        plt.title('EPOCHS VS SSIM')
        
        # function to show the plot
        plt.show()


graphLossResults('T0104_DeepPaint_SmootL1.csv')
graphMetricsResults('T0104_DeepPaint_SmootL1_Metrics.csv')
```
